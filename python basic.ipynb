```python
def addNums(a,b): # add function
    c=a+b
    return c #returning value


c=addNums(2,3) #call Funtion
print(c)
```

```python
def calculateNum(a,b): #  function
    c=a+b
    d=a-b
    return c,d #returning values


c,d=calculateNum(2,3) #call Funtion and catching values
print(c," ",d)
```

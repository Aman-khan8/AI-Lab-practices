```python
import random
#initial population
chromosomes = ["010101","110001","000101","101100"]      #6-bit binary

for i in range(10):
    #step 1: convert to decimal
    decimal_values = []
    for x in chromosomes:
        y = int(x,2)
        decimal_values.append(y)
    
    #step 2: calc fitness values  (specific value/range will be given)
       #steps : step 2.1 -> convert numbers into fixed [0,1] range by dividing each chrmomosome to its highest possible value e.g 2^6 = 64-1 = 63 ,
               # step 2.2 -> now convert these [0,1]   range values to given [-1,1] range by mul/div/sub/add 
               # step 2.3 -> now convert to fixed [0,100] range to get its percentage
    
    fitness_values = []
    for x in decimal_values:
        y= (x/63)*2-1
        z= (y + 1)*50
        fitness_values.append(z)
    
    #step 3 : find best chromosome
    best_fitness = max(fitness_values) #77.77
    best_index = fitness_values.index(best_fitness)
    best_chromo = chromosomes[best_index]
    print(best_chromo)
    
    
    #step 4: Selection (Roulette Wheel)
    selection_probs = []
    for x in fitness_values:
        y = x / sum(fitness_values)
        selection_probs.append(y)
    selected_parents = random.choices(chromosomes, weights=selection_probs,k=2)    # based on weights,K=2 selected parents
    
    #step 5 : Crossover if (single point)
    parent1,parent2 = selected_parents
    cross_point = random.randint(1,len(parent1)-1)
    child1 = parent1[:cross_point]+parent2[cross_point:]
    child2 = parent2[:cross_point]+parent1[cross_point:]
    
    #step 5 : Crossover if (Double point)
    point1,point2 = sorted(random.sample(range(1,len(parent1)-1),2))
    child1 = parent1[:point1]+parent2[point1:point2]+parent1[point2:]
    child2 = parent2[:point1]+parent1[point1:point2]+parent2[point2:]
    
    
    #step 6 : Mutation
    mutation_rate = 0.4
    def mutate(c):
        c_list = list(c)
        for i in range(len(c_list)):
            if random.random() < mutation_rate:
                if c_list[i] == '1':
                    c_list[i]= '0'
                else:
                    c_list[i]='1'
        return ''.join(c_list)
    
    child1 = mutate(child1)
    child2 = mutate(child2)
    
    #step 7 : create new population
    c_w_fitness = list(zip(chromosomes, fitness_values))
    c_w_fitness.sort(key= lambda x: x[1], reverse=True)
    sorted_c =[]
    for x,y in c_w_fitness:
        sorted_c.append(x)
    chromosomes = sorted_c[:-2]+[child1,child2]   #string slicing -> we place child1,child2 in last two indexes
```

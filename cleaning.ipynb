```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
#df.dropna(how="any",inplace=True) #drop all null values if even 1 column is nan
df.dropna(how='all' , inplace=True) #drop all null values if all row is nan
```

```python
for col in df.columns: #columns will give us columns 
    if df[col].dtype=="object":
        df[col]=df[col].fillna(df[col].mode()[0])  # fillna will replace nan values with mode if object 
    else:
        df[col]=df[col].fillna(df[col].median()) #and with median if int or float data type
```

```python
df.tail()
```

```python
df.loc[df["Age"]<0,"Age"]=df["Age"].mean() # short for location 
print(df["Age"])
```

```python
df["Gender"].replace({"M": "Male", "F": "Female", "female": "Female", "male": "Male"}, inplace=True)
print(df["Gender"])
```

```python
df.describe()
```

```python

for col in df.columns:
 if df[col].dtypes!="object":
    Q1 = df['Marks'].quantile(0.25)
    Q3 = df['Marks'].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    df = df[(df['Marks'] >= lower_bound)
     &(df['Marks'] <= upper_bound)]
       
```

```python
le=LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
print(df["Gender"])
```

```python

df.to_csv('Student_Updated.csv' , index=False)
```

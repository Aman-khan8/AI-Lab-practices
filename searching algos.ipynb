```python
import heapq

def ucs(graph, start, goal):
    """
    Perform Uniform Cost Search (UCS) to find the lowest-cost path.
    
    graph: dictionary of dictionaries
           {node: {neighbor: cost, ...}, ...}
    start: starting node
    goal: target node
    
    Returns: (path, total_cost)
    """
    # Priority queue: (cumulative_cost, current_node, path_so_far)
    queue = [(0, start, [start])]
    visited = set()

    while queue:
        cost, node, path = heapq.heappop(queue)

        # If goal is reached, return path and cost
        if node == goal:
            return path, cost

        # Avoid revisiting nodes
        if node in visited:
            continue
        visited.add(node)

        # Explore neighbors
        for neighbor, edge_cost in graph.get(node, []).items():
            if neighbor not in visited:
                heapq.heappush(queue, (cost + edge_cost, neighbor, path + [neighbor]))

    return None, float('inf')  # No path found

# Example usage:
graph = {
    'A': {'B': 2, 'C': 5},
    'B': {'C': 1, 'D': 4},
    'C': {'D': 2, 'E': 3},
    'D': {'E': 1, 'G': 2},
    'E': {'G': 1},
    'G': {}
}

path, total_cost = ucs(graph, 'A', 'G')
print("Path:", path)
print("Total Cost:", total_cost)
```

```python
graph= {
    'S' : {'A' : 1, 'B' : 5},
    'A' : {'C' : 9, 'D' : 5},
    'B' : {'E' : 2, 'F' : 1},
    'C' : {'G' : 1},
    'D' : {'G' : 7},
    'E' : {'G' : 20},
    'F' : {'G' : 10},
    'G' : {}
}


from collections import deque



def bfs(start, goal):
    queue = deque([(start, [start])])  # (node, path)
    visited = set()

    while queue:
        current, path = queue.popleft()

        if current == goal:
            return path

        if current not in visited:
            visited.add(current)

            for neighbor in graph[current]:
                queue.append((neighbor, path + [neighbor]))

    return None

print(bfs('A', 'G'))
```

```python
def dfs(graph, start, goal):
    """
    Depth-First Search (DFS)
    
    graph: dictionary of neighbors (ignore cost)
    start: starting node
    goal: target node
    
    Returns: path from start to goal
    """
    stack = [(start, [start])]  # stack stores (current_node, path_so_far)
    visited = set()
    
    while stack:
        node, path = stack.pop()  # LIFO: last added node is expanded first
        
        if node == goal:
            return path
        
        if node not in visited:
            visited.add(node)
            
            # Add neighbors to stack (DFS goes deeper first)
            # reversed() ensures left-most neighbor is processed first
            for neighbor in reversed(list(graph.get(node, []))):
                if neighbor not in visited:
                    stack.append((neighbor, path + [neighbor]))
    
    return None  


# Corrected graph using lists to preserve order
graph = {
    'S': ['A', 'C', 'B'],  # S has three children
    'A': ['E'],            # A → E
    'E': ['F'],            # E → F
    'F': ['G'],       # F → G, F → C
    'B': ['E', 'D'],       # B → E, B → D
    'C': ['D'],            # C → D
    'D': ['F'],            # D → F
    'G': []                # G has no children
}

path = dfs(graph, 'A', 'G')
print("DFS Path:", path)
```

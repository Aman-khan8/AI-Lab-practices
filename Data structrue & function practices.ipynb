```python
def count_pattern(pattern,mylist):
    count=0;
    for x in range(0,len(mylist)):
    
        if(mylist[x]==pattern[0]):
            smallList=mylist[x:x+len(pattern)]
            if(pattern==smallList):
             count = count + 1

    return count;


mylist=['a','b','c','e','b','a','b','f']
pattern=['a','b']
print("Pattern count is: ",count_pattern(pattern,mylist));
                   
```

```python
def calculator():
    print("Simple Calculator")
    num1 = float(input("Enter first number: "))
    operator = input("Enter operator (+, -, *, /): ")
    num2 = float(input("Enter second number: "))

    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 != 0:
            result = num1 / num2
        else:
            result = "Error! Division by zero."
    else:
        result = "Invalid operator!"

    print("Result:", result)

calculator()
```

```python
def merge_sort_sentence(sentence):
    if len(sentence) <= 1:
        return sentence

    mid = len(sentence) // 2
    left = merge_sort_sentence(sentence[:mid])
    right = merge_sort_sentence(sentence[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i][0] <= right[j][0]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


mess = "I am learning Python"
sentence = mess.split()
sorted_sentence = merge_sort_sentence(sentence)
print(sorted_sentence)
```

```python
def num_conversion(num):
    roman_numerals = [
        ("M", 1000),
        ("CM", 900),
        ("D", 500),
        ("CD", 400),
        ("C", 100),
        ("XC", 90),
        ("L", 50),
        ("XL", 40),
        ("X", 10),
        ("IX", 9),
        ("V", 5),
        ("IV", 4),
        ("I", 1)
    ]
    romanNumber=""
    for key, value in roman_numerals:
        print(key,value)
        while num >= value:
            romanNumber+=key
            num-=value
    return romanNumber


num = int(input("Enter the number you want to convert"))
print(f"{num} is coverted in Roman: {num_conversion(num)}")
```

```python
def twoSum(mylist,target):
    mylist.sort()
    result=[]
    i=0
    j=len(mylist)-1
    while i < j:
        if mylist[i] + mylist[j] > target:
            j=j-1
        elif  mylist[i] + mylist[j] < target:
            i=i+1;
        else:
            result.append(mylist[i])
            result.append(mylist[j])
            break
         
    return result;


mylist=[10,20,10,40,50,60,70]
target=50
print(f"{mylist} has indexes making sum of {target} \n {twoSum(mylist,target)}")
            
        
```

```python
def TwoSum(arr, target):
    result = []
    i = 0
    j = len(arr) - 1

    while i < j:
        s = arr[i] + arr[j]

        if s > target:
            j -= 1
        elif s < target:
            i += 1
        else:
            result.append([arr[i], arr[j]])
            i += 1
            j -= 1

    return result


def threeSum(mylist):
    mylist.sort()
    finallist = []

    for x in range(len(mylist)):
        target = -mylist[x]
        pairs = TwoSum(mylist[x+1:], target)

        for pair in pairs:
            finallist.append([mylist[x]] + pair)

    return finallist


mylist = [-25, -10, -7, -3, 2, 4, 8, 10]
print(threeSum(mylist))
```

```python
def matrix_add():
  A = [[1,2],[3,4]]
  B = [[5,6],[7,8]]
  result = [[0,0],[0,0]]
  for i in range(len(A)):
    for j in range(len(A[0])):
      result[i][j] = A[i][j] + B[i][j]
  return result
res = matrix_add()
print(res)
```

```python
def matrix_mul():
  A = [[1,2],[3,4]]
  B = [[5,6],[7,8]]
  result = [[0,0],[0,0]]
  for i in range(len(A)):
    for j in range(len(B[0])):
      for k in range(len(B)):
        result[i][j] += A[i][k] * B[k][j]
  return result
res = matrix_mul()
print(res)
```

```python
def dups(mylist):
    duplicates = []
    seen = []

    for num in mylist:
        if num in seen and num not in duplicates:
            duplicates.append(num)
        else:
            seen.append(num)

    return duplicates


# Input
lst = list(map(int, input("Enter list: ").split()))
print("Duplicates:", dups(lst))
```
